```python
import pandas as pd
from ftfy import fix_text

# get the data
data = pd.read_csv("all_items_gplay.csv")
```

```python
# fix mojibake
for col in data.select_dtypes(include='object'):
    data[col] = data[col].astype(str).apply(fix_text)
    
    
import re
# find all the parenthesis and content in the game id and remove the parenthesis
find_parenthesis = re.compile(r"\(([^)]*)\)")



# apply the regex to the 'game_id' column and extract the first match
data['game_name_from_item_id'] = data['item_id'].apply(lambda x: find_parenthesis.search(x).group(0) if find_parenthesis.search(x) else "")
```
